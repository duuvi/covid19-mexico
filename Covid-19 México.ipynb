```python
valores = datos_FR.resample("D").count().cumsum().values
valores
```

```python
dias = range(1, len(valores) + 1)
```

```python
plt.plot(dias, valores)
plt.xlabel("Dias transcurridos desde el primer caso")
plt.ylabel("Incrementos acumulados")
```

## Ajuste y Pronóstico

```python
def logisticModel(x, a, b, c):
    return c / ( 1 + np.exp( - (x - b) / a ) )
```

```python
x = np.arange(1, len(valores) + 1)
```

```python
y = valores.flatten()
```

```python
fit = curve_fit(logisticModel, x, y, p0 = [1, 10, 1000])
fit
```

```python
parametros, matriz_covarianza = fit
```

```python
a, b, c = parametros
```

```python
x_model = np.linspace(0, len(valores))
```

```python
plt.plot(x, y, "ok", label = "Datos")
plt.plot(x_model, logisticModel(x_model, a, b, c), "r", label = "Ajuste logístico")
plt.xlabel("Dias transcurridos desde el primer caso")
plt.ylabel("Incrementos acumulados")
plt.legend()
plt.grid()
plt.show()
```

```python
# Predicción a una semana
x_model = np.linspace(0, len(valores) + 7)
```

```python
plt.plot(x, y, "ok", label = "Datos")
plt.plot(x_model, logisticModel(x_model, a, b, c), "r", label = "Ajuste logístico")

plt.plot( [ x_model[-1], x_model[-1] ], [ 0, logisticModel( x_model[-1], a, b, c ) ], "--b", label = "Pronóstico dentro de una semana" )

plt.xlabel("Dias transcurridos desde el primer caso")
plt.ylabel("Incrementos acumulados")
plt.legend()
plt.grid()
plt.show()
```
